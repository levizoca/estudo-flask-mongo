# Python

```python
### Importando bibliotecas

import pandas as pd
```

```python
### Lendo arquivo e renomeando colunas

filename = 'K3241.K03200Y9.D30311.ESTABELE.csv'
columns = ['CNPJ_BÁSICO',
           'CNPJ_ORDEM ',
           'CNPJ_DV',
           'IDENTIFICADOR_MATRIZ/FILIAL',
           'NOME_FANTASIA',
           'SITUAÇÃO_CADASTRAL',
           'DATA_SITUAÇÃO_CADASTRAL',
           'MOTIVO_SITUAÇÃO_CADASTRAL',
           'NOME_DA_CIDADE_NO_EXTERIOR',
           'PAIS',
           'DATA_DE_INÍCIO_ATIVIDADE ',
           'CNAE_FISCAL_PRINCIPAL',
           'CNAE_FISCAL_SECUNDÁRIA',
           'TIPO_DE_LOGRADOURO',
           'LOGRADOURO',
           'NÚMERO',
           'COMPLEMENTO',
           'BAIRRO',
           'CEP',
           'UF',
           'MUNICÍPIO',
           'DDD_1',
           'TELEFONE_1',
           'DDD_2',
           'TELEFONE_2',
           'DDD_DO_FAX',
           'FAX',
           'CORREIO_ELETRÔNICO',
           'SITUAÇÃO_ESPECIAL',
           'DATA_DA_SITUAÇÃO_ESPECIAL']
df = pd.read_csv(filename, sep=';', names = columns, encoding='latin-1')
```

```python
### Vendo numero de linhas e colunas

df.shape
```

```python
### Criando colunas CNPJ COMPLETO e TELEFONE

df['CNPJ_COMPLETO'] = df['CNPJ_BÁSICO'].astype(str) + df['CNPJ_ORDEM '].astype(str) + df['CNPJ_DV'].astype(str)
df['TELEFONE'] = df['DDD_1'].astype(str) + "/" + df['TELEFONE_1'].astype(str)
```

```python
### separado colunas

dfreg = df.sample(10)
dfreg['TELEFONE'] = dfreg['TELEFONE'].str.replace('.0', '')
dfreg.sample(10)
```

```python
### Criando pivot_table apenas com as colunas que serão utilizadas

registros = dfreg.pivot_table(
    index='CNPJ_COMPLETO',
    values=['NOME_FANTASIA', 'CEP', 'TELEFONE', 'CORREIO_ELETRÔNICO'],
    aggfunc='first'
)
registros.sample(10)
```

```python
### Salvando arquivo em csv com apenas 10 registros

registros.sample(10).to_csv('estabelecimentos10.csv')
```

```python
### Transformando arquivo csv em json

df = pd.read_csv('estabelecimentos10.csv')
df = df.fillna('None')
estabelecimentos_data = df.to_dict(orient = "records")
estabelecimentos_data
```

# Elastic Search

```python
### Importando bibliotecas para conexão com o Elasticsearch

import elasticsearch
print(elasticsearch.__version__)
```

```python
### Conectando ao Elasticsearch

es = elasticsearch.Elasticsearch([{'host': 'localhost', 'port': 9200}])
resp = es.info()
```

```python
### Inserindo dados no Elasticsearch

estabelecimentos_data = df.to_dict(orient = "records")
for data in estabelecimentos_data: 
  es.index(index='estabelecimentos', document=data)
```

```python
### Buscando dados no Elasticsearch

res = es.search(index="estabelecimentos", body={'size':1000, "query": {"match_all": {}}})
print("Got %d Hits:" % res['hits']['total']['value'])

for hit in res['hits']['hits']:
    print("%(CNPJ_COMPLETO)s %(NOME_FANTASIA)s: %(TELEFONE)s" % hit["_source"])
```
